## Sumamrizations

```python
from transformers import pipeline

summarizer = pipeline("summarization")

text = """The quick brown fox jumps over the lazy dog. This sentence is often used to
demonstrate the use of all the letters in the English alphabet. It is a pangram,
which means it contains every letter at least once."""

print(summarizer(text, max_length=50, min_length=25, do_sample=False))
```

## Translations

```python
translator_ko = pipeline(
    "translation",
    model="Helsinki-NLP/opus-mt-en-ko"
)

print(translator_ko("i hate you"))
```

```python
pipe = pipeline("translation", model="Helsinki-NLP/opus-mt-en-fr")
sentence = "How you doing?"
print(pipe(sentence))
```

## Emotions

```python
from transformers import pipeline
classifier = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base", return_all_scores=True)
classifier("Im exhuasted but Alhamdulillah")
classifier("I just dont wanna do anything today!")
classifier("There are alot of things ot be done today and still i feel im not that much active.")
```

## Text generation with gpt 

```python
from transformers import GPT2DoubleHeadsModel, GPT2Tokenizer

tokenizer = GPT2Tokenizer.from_pretrained("gpt2-large")
model = GPT2DoubleHeadsModel.from_pretrained("gpt2-large", pad_token_id=tokenizer.eos_token_id)
sentence = "Why we forgot what is actually the purpose of our life and what is our actual destination?"
inputs = tokenizer.encode(sentence, return_tensors="pt")
inputs

for i in inputs[0]:
    print(f"the word is {tokenizer.decode(i)}----------> id:{i}")
outputs = model.generate(inputs, max_length=100, no_repeat_ngram_size=2,  num_beams=5)
outputs
print(tokenizer.decode(outputs[0], skip_special_tokens=True))
    
```
